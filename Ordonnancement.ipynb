```python

nbr_encap = 8
nbr_etiq = 2
nbr_simu = 3
max_agent = 8
agents = [i+1 for i in range(max_agent)]
means_bout = []
for nbr_agent in range(1,max_agent + 1):
    X,Y = multi_simu(nbr_encap,nbr_etiq,nbr_agent,nbr_simu)
    mean_bout = sum(Y)/len(Y)
    means_bout.append(mean_bout/nbr_agent)
plt.plot(agents,means_bout)
    
```
